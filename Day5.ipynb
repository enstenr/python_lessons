# Python Sets

```python
# lets pull in country from list
from pandas import  * 
data = read_csv("../Documents/countries.csv")
country_set = set(data['Country'])

# sets are unordered unchangeable , unique 

st = { 1, True, 0, False}
st
```

```python
# using set constructor to create a set 
st = set(("Rajesh"))
st
```

```python
# Accessing set 
for x in st : 
    print(x)  
```

```python
'R' in st
```

```python
# once added, we cannot change an item from set
st.add ('S')
st
```

```python
# Add two sets using update method 
country_set.update(st)
country_set
```

```python
# remove item from set 
country_set.remove(1)
country_set
```

```python
country_set.discard(1)
country_set
```

```python
country_set.clear()
country_set

del country_set
country_set
```

```python
# union , intersection, difference, symmetric difference 
st1 = {x for x in range(10)}
st2 = { x for x in range (5, 15)}

```

```python
st1.union(st2)
```

```python
st1.intersection(st2)
```

```python
st1.difference(st2)
```

```python
st1.symmetric_difference(st2)
```
