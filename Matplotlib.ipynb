```python

import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace( -4, 4, 20 )
y = x*x 
y2 = x*x*x 

plt.plot(x,y)
plt.plot(x,y2)

plt.show()
```

```python
# Plotting scattered points

import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(-4, 4, 15 ) 
y = x*x
y2 = x*x*x 

plt.plot(x,y,'*')
plt.show()
```

```python
# Scattered plot 
import numpy as np 
import matplotlib.pyplot as plt 

x = np.random.random(50)
y = np.random.random(50)

sz = np.random.randint(2,260,50)
plt.scatter(x,y, s=sz)
```

```python
import calendar
import numpy as np 
import matplotlib.pyplot as plt

#help(calendar)
months= [calendar.month_name[x].upper() for x in range( 1, 13)]
sales_comp = np.random.randint(1,100,12)
competitor_sales = np.random.randint(1,100,12)

plt.bar(months,sales_comp)
plt.bar(months,competitor_sales, bottom = sales_comp)
plt.show()
```
