```python
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline

```

```python
data = np.array([1,2,3,4])
data
plt.plot(data)
plt.show()
```

```python
x = np.linspace(-4,4,20)
y = x*x
y2 = x**3
plt.plot(x,y,'ro')
plt.plot(x,y2,'x')
plt.show()
```

```python
# bubbleplot 
plt.figure(figsize=(9,3))
x = np.random.random(50)
y = np.random.random(50)
sz = np.random.randint(2,200,50)
plt.scatter(x,y,s=sz,c='r')
plt.show()
```
