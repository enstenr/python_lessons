```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='day',size='size',sizes =(10,200))
```

```python
sns.catplot(x='day',y='total_bill',kind='swarm',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips,hue='sex',split=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
flights = sns.load_dataset("flights").pivot("month", "year", "passengers")

# Create a figure with multiple subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))

# Draw the first heatmap on the first Axes
sns.heatmap(flights, annot=True, fmt='d', linewidths=0.5, cmap='YlGnBu', ax=ax1)
ax1.set_title('Heatmap 1: YlGnBu')

# Draw the second heatmap on the second Axes with a different color palette
sns.heatmap(flights, annot=True, fmt='d', linewidths=0.5, cmap='coolwarm', ax=ax2)
ax2.set_title('Heatmap 2: coolwarm')

plt.show()
```
